```python
import matplotlib.pyplot
```

```python
%matplotlib inline
```

```python
data = numpy.loadtxt(fname='data/080167500_Guadalupe_SpringBranch_DailyMeanQ.csv', delimiter=',')
```

```python
print(data)
```

```python
print(numpy.mean(data, axis = 0))
```

```python
print(numpy.mean(data, axis = 0))
print(numpy.min(data, axis = 0))
print(numpy.max(data, axis = 0))
print(numpy.max(data, axis = 0)-numpy.min(data, axis = 0))
```

```python
meanQ = numpy.mean(data, axis=0)
minQ = numpy.min(data, axis=0)
maxQ = numpy.max(data, axis=0)
rangeQ = maxQ - minQ
print(meanQ)
print(minQ)
print(maxQ)
print(rangeQ)
```

```python
image = matplotlib.pyplot.imshow(data)
matplotlib.pyplot.show()
```

```python
plt = matplotlib.pyplot
```

```python
meanQ_plot = plt.plot(meanQ)
minQ_plot = plt.plot(minQ)
maxQ_plot = plt.plot(maxQ)
rangeQ_plot = plt.plot(rangeQ)
```

```python
fig1 = plt.figure(figsize=(14.0, 3.0))

meanQ_subplot = fig1.add_subplot(1, 4, 1)
minQ_subplot = fig1.add_subplot(1, 4, 2)
maxQ_subplot = fig1.add_subplot(1, 4, 3)
rangeQ_subplot = fig1.add_subplot(1, 4, 4)

meanQ_subplot.set_ylabel('mean')
meanQ_subplot.plot(meanQ)
minQ_subplot.set_ylabel('min')
minQ_subplot.plot(minQ)
maxQ_subplot.set_ylabel('max')
maxQ_subplot.plot(maxQ)
rangeQ_subplot.set_ylabel('range')
rangeQ_subplot.plot(rangeQ)

fig1.tight_layout()

plt.show()
```
