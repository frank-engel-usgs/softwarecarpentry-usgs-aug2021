```python
import numpy
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = numpy.genfromtxt(fname='080167500_Guadalupe_SpringBranch_DailyMeanQ.csv', delimiter=',')
data[data == 0] = numpy.nan
print(data)
```

```python
Q_mean = numpy.nanmean(data, axis=0) # takes the mean of each col
```

```python
print(Q_mean.round(2))
```

```python
Q_min = numpy.nanmin(data, axis=0) # takes the min of each col
```

```python
print(Q_min.round(1)) # does not ignore Zero's
```

```python
Q_max = numpy.nanmax(data, axis=0) # takes the max of each col
```

```python
print(Q_max)
```

```python
Q_range = Q_max - Q_min
```

```python
print(Q_range)
```

```python
fig = plt.figure(figsize=(15,4))
```

```python
axes1 = fig.add_subplot(1, 4, 1)
axes2 = fig.add_subplot(1, 4, 2)
axes3 = fig.add_subplot(1, 4, 3)
axes4 = fig.add_subplot(1, 4, 4)
```

```python
axes1.set_ylabel('average', fontweight='bold', fontsize=12)
axes1.plot(Q_mean)
```

```python
axes2.set_ylabel('max', fontweight='bold', fontsize=12)
axes2.plot(Q_max)
```

```python
axes3.set_ylabel('min', fontweight='bold', fontsize=12)
axes3.plot(Q_min)
```

```python
axes4.set_ylabel('range', fontweight='bold', fontsize=12)
axes4.plot(Q_range)
```

```python

fig.tight_layout()
```
